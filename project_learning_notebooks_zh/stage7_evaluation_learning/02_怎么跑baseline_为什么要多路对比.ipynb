# 02 怎么跑 baseline：为什么要多路对比

这一节解释你刚才最困惑的点：为什么要跑 baseline，为什么要生成多路回答，为什么要设置环境变量。

## 1. baseline 是什么

baseline 就是“训练前的参照物”。没有 baseline，就不知道微调后是变好了还是只是感觉变好了。

在这个项目里，至少有三种 baseline：

```text
base Qwen：原始模型能力。
custom-SFT v3：之前技术任务里最稳的 adapter。
DPO v6：之前最好的 DPO 实验 artifact，但不是默认推荐。
```

跑 baseline 的目的：先知道它们在安全帮助任务上分别差在哪里。

## 2. 为什么要多路回答

多路回答不是“自己评估自己”。它是让不同模型版本参加同一场考试：

```text
同一批 prompt
同一套生成参数
不同模型版本回答
同一套评分规则 / 人工 rubric 评分
```

这样你能回答面试官的问题：

```text
base 模型哪里差？
之前的技术 SFT 是否已经影响安全回答？
DPO artifact 是否带来帮助或副作用？
新训练后是否真的超过这些 baseline？
```

模型负责生成，评测集和评分器负责评价。

## 3. 为什么设置 HF_HOME

命令：

```powershell
$env:HF_HOME=(Resolve-Path -LiteralPath ".hf_cache").Path
$env:HF_HUB_DISABLE_SYMLINKS_WARNING="1"
```

意思是：

```text
HF_HOME：让 Hugging Face 模型缓存放在项目的 .hf_cache 目录，避免系统目录混乱。
HF_HUB_DISABLE_SYMLINKS_WARNING：Windows 上关闭符号链接警告，减少无关噪音。
```

这和评估思想无关，是为了让本地加载模型更稳定。

## 4. 为什么 temperature=0

评估时通常先用确定性生成：

```text
temperature=0
同一个 prompt 尽量得到同一个回答
方便比较不同模型版本
```

如果 temperature 很高，模型每次回答都不一样，评估会很难复现。

## 5. 这一节先不要求你跑

等你读完 00-04，我们再一起跑 baseline。现在你只需要理解：

```text
baseline = 训练前参照
多路回答 = 同一张考卷给多个模型版本做
评分器/人工 rubric = 外部裁判
temperature=0 = 让考试可复现
HF_HOME = 本地缓存稳定
```
