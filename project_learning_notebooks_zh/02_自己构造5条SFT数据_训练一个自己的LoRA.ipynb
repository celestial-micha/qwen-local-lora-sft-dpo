# 02 自己构造 5 条 SFT 数据：训练一个自己的 LoRA

今天只解决一个问题：

> 我怎么自己写几条数据，然后让项目用这些数据训练一个 LoRA adapter？

5 条数据只能用于理解流程，不能指望模型质量好。它的意义是让你掌握从数据到训练命令的完整手感。

````python
from pathlib import Path
import json, os, subprocess, sys, textwrap

PROJECT_ROOT = Path.cwd()
if not (PROJECT_ROOT / "scripts").exists():
    PROJECT_ROOT = Path(r"D:/coding/codex备份/qwen-local-lora-sft-dpo学习版")

print("项目根目录:", PROJECT_ROOT)

def read_text(rel):
    return (PROJECT_ROOT / rel).read_text(encoding="utf-8", errors="replace")

def show_file(rel, start=1, end=80):
    lines = read_text(rel).splitlines()
    print(f"--- {rel}:{start}-{min(end, len(lines))} ---")
    for i in range(start, min(end, len(lines)) + 1):
        print(f"{i:03d}: {lines[i-1]}")

def find_lines(rel, keyword, context=4):
    lines = read_text(rel).splitlines()
    hits = [i for i, line in enumerate(lines, start=1) if keyword in line]
    if not hits:
        print("没有找到:", keyword)
        return
    for hit in hits:
        show_file(rel, max(1, hit-context), min(len(lines), hit+context))
        print()

def read_jsonl(rel, n=3):
    rows = []
    with (PROJECT_ROOT / rel).open("r", encoding="utf-8") as f:
        for line in f:
            line = line.strip()
            if line:
                rows.append(json.loads(line))
            if len(rows) >= n:
                break
    return rows

def count_jsonl(rel):
    with (PROJECT_ROOT / rel).open("r", encoding="utf-8") as f:
        return sum(1 for line in f if line.strip())
````

## 1. 项目支持什么数据格式？

如果你自己写 SFT 数据，有两种路线。

### 路线 A：直接写 Qwen chat JSONL

```json
{"messages":[
  {"role":"system","content":"..."},
  {"role":"user","content":"..."},
  {"role":"assistant","content":"..."}
]}
```

### 路线 B：写 Alpaca 格式，再用 `prepare_data.py` 转换

```json
{"instruction":"...", "input":"", "output":"..."}
```

当前最直接、最容易理解的是路线 A。

## 4. 如果要真的写文件，写到 learning_try 目录

默认不写。你确认想练习时，把 `WRITE_FILES=True`。

输出会在：

```text
data/learning_try/tiny_sft_train.jsonl
data/learning_try/tiny_sft_eval.jsonl
```

不覆盖正式 `data/processed/`。

````python
WRITE_FILES = False
if WRITE_FILES:
    out_dir = PROJECT_ROOT / "data" / "learning_try"
    out_dir.mkdir(parents=True, exist_ok=True)
    train_path = out_dir / "tiny_sft_train.jsonl"
    eval_path = out_dir / "tiny_sft_eval.jsonl"
    train_rows = my_rows[:4]
    eval_rows = my_rows[4:]
    for path, rows in [(train_path, train_rows), (eval_path, eval_rows)]:
        with path.open("w", encoding="utf-8") as f:
            for row in rows:
                f.write(json.dumps(row, ensure_ascii=False) + "\n")
        print("写入:", path, "行数:", len(rows))
else:
    print("默认不写文件。理解格式后再改 WRITE_FILES=True。")
````

## 5. 训练命令长什么样？

这里不默认训练。你先看命令如何连接到 `train_sft_lora.py` 的参数。

````python
cmd = [
    sys.executable, "scripts/train_sft_lora.py",
    "--train_file", "data/learning_try/tiny_sft_train.jsonl",
    "--eval_file", "data/learning_try/tiny_sft_eval.jsonl",
    "--output_dir", "outputs/learning_try_tiny_sft_lora",
    "--max_length", "256",
    "--batch_size", "1",
    "--grad_accum", "1",
    "--epochs", "1",
    "--lr", "2e-4",
    "--logging_steps", "1",
    "--eval_steps", "5",
    "--save_steps", "5",
    "--report_to", "none",
    "--local_files_only",
]
print(" ".join(cmd))
RUN_TINY_TRAINING = False
if RUN_TINY_TRAINING:
    result = subprocess.run(cmd, cwd=PROJECT_ROOT, text=True, capture_output=True, encoding="utf-8", errors="replace")
    print(result.stdout)
    if result.stderr:
        print(result.stderr)
else:
    print("默认不训练。要训练，先 WRITE_FILES=True 写数据，再 RUN_TINY_TRAINING=True。")
````
